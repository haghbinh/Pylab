در پایتون، متدهای جادویی (<b>Magic Methods</b>) متدهایی هستند که با دو underscore قبل و بعد از نام تعریف می‌شوند و رفتارهای داخلی اشیاء را کنترل می‌کنند.  
با استفاده از این متدها می‌توان عملگرها را برای کلاس‌های خود بازتعریف کرد (<b>Operator Overloading</b>).
</p>

<h2 dir="rtl" style="text-align: justify;">۱. مثال __str__ و __repr__</h2>


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"Product: {self.name}, Price: {self.price}"

    def __repr__(self):
        return f"Product('{self.name}', {self.price})"

p = Product("Laptop", 1500)
print(p) # __str__ فراخوانی می‌شود
```

```python
print(repr(p)) # __repr__ فراخوانی می‌شود
```

```python
class Rectangle:
	def __init__(self, l , w):
		self.length = l
		self.breadth = w
	def __repr__(self):
		return f' Rectangle with length {self.length} and breadth {self.breadth}'
	def Calc_Area(self, obj):
		print(' Length = ',obj.length)
		print(' Breadth = ',obj.breadth)
		return obj.length * obj.breadth

Obj1 = Rectangle(10,20)
print('The area of Rectangle is ', Obj1.Calc_Area(Obj1))

```

<h2 dir="rtl" style="text-align: justify;">۲. بازتعریف عملگر جمع</h2> <p dir="rtl" style="text-align: justify;"> می‌توانیم رفتار عملگر <code>+</code> را برای کلاس خود تعریف کنیم. </p>

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __add__(self, other):
        # جمع قیمت دو محصول
        return self.price + other.price

p1 = Product("Laptop", 1500)
p2 = Product("Phone", 800)
total = p1 + p2
print(total)  
```

# Special Methods for Arithmetic Operations
A programmer can overload any arithmetic operation by implementing the corresponding special method. 

<figure style="text-align: center;">
  <picture>
    <img loading="lazy" src="images/special_methods.png" width="1000" >
  </picture>
</figure>

```python
class OprOverloadingDemo:
	def __init__(self,X):
	    self.X = X
	def __add__(self,other):
		print(' The value of Ob1 =',self.X)
		print(' The value of Ob2 =',other.X)
		print(' The Addition of two objects is:',end='')
		return ((self.X+other.X))

Ob1 = OprOverloadingDemo(20)
Ob2 = OprOverloadingDemo(30)
Ob3 = Ob1 + Ob2
print(Ob3)
```


<h2 dir="rtl" style="text-align: justify;">۳. متدهای مقایسه</h2> <p dir="rtl" style="text-align: justify;"> می‌توانیم عملگرهای مقایسه را بازتعریف کنیم: </p>

```python
class Test:
	def __init__(self, x , y):
		self.a = x
		self.b = y
	def __eq__(self, obj):
		if(obj.a == self.a and obj.b == self.b):
			return True
		else:
			return False


Obj1 = Test(10,20)
Obj2 = Test(10,20)
Obj3 = Test(12,90)
print(' Obj1 == Obj2 ',Obj1==Obj2)
print(' Obj1 == Obj3 ',Obj1==Obj3)
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __lt__(self, other):  # less than
        return self.price < other.price

    def __eq__(self, other):  # equal
        return self.price == other.price

p1 = Product("Laptop", 1500)
p2 = Product("Phone", 800)
print(p1 < p2)  
```

```python
print(p1 == p2) 
```

# Special Methods for Comparing Types
Comparison is not strictly done on numbers. It can be made on various types, such as list, string and even on dictionaries.

<figure style="text-align: center;">
  <picture>
    <img loading="lazy" src="images/comparing_types.png" width="1000" >
  </picture>
</figure>

```python
class CmpOprDemo:
	def __init__(self,X):
		self.X = X
	def __lt__(self,other):
		print(' The value of Ob1 =',self.X)
		print(' The value of Ob2 =',other.X)
		print(' Ob1 < Ob2 :',end='')
		return self.X <other.X
	def __gt__(self,other):
		print(' Ob1 > Ob2 :',end='')
		return self.X > other.X
	def __le__(self,other):
		print(' Ob1 <= Ob2 :',end='')
		return self.X <= other.X

Ob1 = CmpOprDemo(20)
Ob2 = CmpOprDemo(30)
print( Ob1 < Ob2 )
print( Ob1 > Ob2 )
print( Ob1 <= Ob2 )
```
